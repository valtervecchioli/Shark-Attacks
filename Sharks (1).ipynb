```python
import pandas as pd
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"
sharkattacks = pd.read_excel(url)
display(sharkattacks.head(3))
sharkattacks.shape
```

```python
#Standardizing column names
sharkattacks.columns = [x.strip() for x in sharkattacks.columns]
sharkattacks.columns = [x.lower().replace(' ', '_') for x in sharkattacks.columns]
sharkattacks = sharkattacks.applymap(lambda x:x.lower() if isinstance(x,str) else x)

sharkattacks.columns
```

```python
#Cleaning "YEAR" column
sharkattacks["year"].isnull().sum()
year_mode=int(sharkattacks["year"].mode())
sharkattacks["year"].fillna(year_mode,inplace=True)
sharkattacks["year"].isnull().sum()
sharkattacks["year"].unique()
sharkattacks["year"]=sharkattacks["year"].map({x:year_mode if x==0 or x==5 or x==77 or x==1000 else x for x in sharkattacks["year"]})
sharkattacks["year"]=sharkattacks["year"].map({x:year_mode if x==1500 else x for x in sharkattacks["year"]})
sharkattacks["year"].value_counts()
sharkattacks["year"] = sharkattacks["year"].astype(int)
```

```python
#Cleaning "COUNTRY" column
sharkattacks['country'] = sharkattacks['country'].astype(str)
sharkattacks['country'] = sharkattacks['country'].map(lambda x: x.upper())
sharkattacks["country"] = sharkattacks["country"].fillna("UNKNOWN")
sharkattacks["country"] = sharkattacks["country"].str.strip()
sharkattacks["country"] = sharkattacks["country"].replace({"UKNOWN": "UNKNOWN",'nan':"UNKNOWN",'NAN' : "UNKNOWN"})
sharkattacks["country"].nunique()
```

```python
#Cleaning "LOCATION" column
sharkattacks["location"].isnull().sum()
sharkattacks["location"].fillna("UNKNOWN",inplace=True)
sharkattacks["location"].isnull().sum()
sharkattacks["location"]=sharkattacks["location"].apply(lambda x:x.title())
```

```python
#Cleaning "ACTIVITY" column
sharkattacks['activity'] = sharkattacks['activity'].astype(str)
sharkattacks['activity'] = sharkattacks['activity'].map(lambda x: x.title())
sharkattacks["activity"] = sharkattacks["activity"].fillna("UNKNOWN")
sharkattacks["activity"] = sharkattacks["activity"].str.strip()
sharkattacks["activity"] = sharkattacks["activity"].dropna().apply(lambda x: "Fishing" if 'fish' in x else x)
sharkattacks["activity"] = sharkattacks["activity"].dropna().apply(lambda x: "Fishing" if 'Fishing' in x else x)
sharkattacks["activity"] = sharkattacks["activity"].dropna().apply(lambda x: "Fishing" if 'Fish' in x else x)
sharkattacks["activity"] = sharkattacks["activity"].dropna().apply(lambda x: "Fishing" if 'fishing' in x else x)
sharkattacks["activity"] = sharkattacks["activity"].replace({'NAN' : "UNKNOWN", 'Nan' : "UNKNOWN"})
```

```python
#Cleaning "SEX" column
sharkattacks['sex'] = sharkattacks['sex'].map({'n': 'm', 'ill': 'm', 'm x 2':'.', '.': '.', 'm': 'm', 'f': 'f'})
sharkattacks['sex'].fillna('m', inplace=True)
sharkattacks= sharkattacks.drop(index=3580)
sharkattacks= sharkattacks.drop(index=6069)
sharkattacks= sharkattacks.drop(index=5568)
```

```python
#Cleaning "STATE" column
sharkattacks['state'] = sharkattacks['state'].astype(str)
sharkattacks['state'] = sharkattacks['state'].map(lambda x: x.title())
sharkattacks["state"] = sharkattacks["state"].fillna("UNKNOWN")
sharkattacks["state"] = sharkattacks["state"].str.strip()
sharkattacks["state"] = sharkattacks["state"].replace({"UKNOWN": "UNKNOWN",'nan':"UNKNOWN",'NAN' : "UNKNOWN"})
```

```python
#Cleaning "TYPE" column:
sharkattacks["type"].isnull().sum()
sharkattacks["type"].fillna("unidentified", inplace = True)
sharkattacks["type"] = sharkattacks["type"].apply(lambda x: x.strip())
sharkattacks ["type"] = sharkattacks ["type"].replace("?","UNKNOWN")
sharkattacks ["type"] = sharkattacks ["type"].replace("boat","UNKNOWN")
```

```python
#Cleaning "NAME" column:
sharkattacks["name"].isnull().sum()
sharkattacks["name"].fillna("UNKNOWN",inplace=True)
sharkattacks["name"].isnull().sum()
sharkattacks["name"].value_counts()
sharkattacks["name"]=sharkattacks["name"].replace("2 males","two males")
sharkattacks["name"]=sharkattacks["name"].apply(lambda x:x.title())
```

```python
#Cleaning "INJURY" column:
sharkattacks["injury"].isnull().sum()
sharkattacks["injury"].fillna("UNKNOWN", inplace = True)
sharkattacks["injury"].isnull().sum()
sharkattacks["injury"] = sharkattacks["injury"].apply(lambda x: x.strip())
```

```python
#Cleaning "SOURCE" column:
sharkattacks['source'] = sharkattacks['source'].astype(str)
sharkattacks["source"].isnull().sum()
sharkattacks["source"].fillna("UNKNOWN", inplace = True)
sharkattacks["source"].isnull().sum()
sharkattacks["source"] = sharkattacks["source"].apply(lambda x: x.strip())
```

```python
#Cleaning "AGE" column:
sharkattacks['age'] = sharkattacks['age'].map({x: str(x).strip(' ') for x in sharkattacks['age']})
sharkattacks['age'] = sharkattacks['age'].map({x:'UNKNOWN' if len(x)>2 else x for x in sharkattacks['age']})
sharkattacks['age'] = sharkattacks['age'].map({x:'UNKNOWN' if not(str(x).isdigit()) else x for x in sharkattacks['age']})
sharkattacks['age'] = sharkattacks['age'].map({x: 28 if x == 'UNKNOWN' else x for x in sharkattacks['age']})
sharkattacks['age'] = sharkattacks['age'].map({x: int(x) if isinstance(x,str) else x for x in sharkattacks['age']})
```

```python
#Dropping weird columns:
sharkattacks.drop(columns=["unnamed:_11","unnamed:_21","unnamed:_22","pdf","href_formula","href","case_number","case_number.1","original_order"], inplace=True)
```
